```python
# Function to compute the cube interval (L)
# Input:
#   data => input data (DataFrame)
#   m    => number of spatial grid (list with length two)
#   m_t  => number of temporal grid (int)
def compute_L(data, m, m_t):
    d1 = abs(data['Lat'].max()-data['Lat'].min())
    d2 = abs(data['Lon'].max()-data['Lon'].min())
    d3 = abs(data['Time'].max()-data['Time'].min())
    L = [d1/m[0], d2/m[1], np.ceil(d3/m_t)]
    
    return L
```

```python
# Function to divide the grid and insert the data into corresponding grid
# Input:
#   data => input data (DataFrame)
#   L    => grid interval (list with length 3)
def partitioning(data, L):
    # create coordinate for the cube
    gridCoord = np.floor(data.loc[:,['Lat','Lon','Time']]/np.tile(L, (data.shape[0], 1)))
    
    # join the grid coordinate and the other attributes
    gridCoord = pd.concat([gridCoord, data.iloc[:,3:]], axis=1)

    # create cube hash table
    cube = [] # save the cube to list
    
    grid_temp = gridCoord # copy the grid cube
    
    for i in range(gridCoord.shape[0]): # for each grid cube coordinate
        # find all objects in the current grid
        objInGrid = grid_temp.loc[(grid_temp['Lat']==gridCoord.loc[i,'Lat']) & (grid_temp['Lon']==gridCoord.loc[i,'Lon']) & (grid_temp['Time']==gridCoord.loc[i,'Time'])]
        if objInGrid.shape[0] != 0: # if there is one object or more in the grid
            grid = {} # create grid cube dictionary that includes Grid coordinates and Object in the grid cube
            grid['Grid'] = list(gridCoord.loc[i, ['Lat','Lon','Time']])
            grid['Data'] = objInGrid
            # add the grid cube dictionary to the cube
            cube.append(grid)
            # delete the objects that have been added to the cube
            grid_temp = grid_temp.drop(objInGrid.index)
    return cube, gridCoord
```

```python
# Function to compute the radius
def compute_r(L,lambda_):
    r1 = min(L[:2])/2
    r2 = max(L[:2])/2
    
    if lambda_ > 1 and lambda_ < 0:
        lambda_ = abs(1/lambda_)
    elif lambda_ > 1:
        lambda_ = 1/lambda_
    elif lambda_ < 0:
        lambda_ = abs(lambda_)
        
    r = lambda_*r1 + (1-lambda_)*r2 - 0.0001
    
    return r
```
